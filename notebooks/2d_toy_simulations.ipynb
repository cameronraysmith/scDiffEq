## Generating toy simulations

This notebook focuses on two datasets simulated from sets of dynamical, non-linear equations.

#### Import libraries

```python
%load_ext nb_black

import scdiffeq as sdq
```

#### Example 1: spiral dynamic system with a single attractor: 

$$\textbf{F(x)}=\begin{equation}
\left\{ 
  \begin{aligned}
    \frac{dx}{dt} = \textbf{F}_{1}(x, y) = -x + y^{2}\\
    \frac{dy}{dt} = \textbf{F}_{2}(x, y) = -y - xy\\
  \end{aligned}
  \right.
\end{equation}$$

#### Instantiate the `GenericSimulator` class and perform the simulation.

```python
Simulation = sdq.data.GenericSimulator()
Simulation.set_initial_conditions_sampling_distribution("normal")
Simulation.get_initial_conditions(
    loc=0, scale=2, size=[5000, 2], plot=True
)  # i.e., normal(loc=0.0, scale=1.0, size=None)
Simulation.create_time_vector(
    time_span=0.35,
    n_samples=100,
)
Simulation.simulate_ODE("parabola_2d")
```

#### Example 2: two-dimensional dynamical system with four attractors:

<br></br>

$$\textbf{F(x)}=\begin{equation}
\left\{ 
  \begin{aligned}
    \frac{dx}{dt} = \textbf{F}_{1}(x, y) = -1 + 9x - 2x^{3} + 9y - 2y^{3}\\
    \frac{dy}{dt} = \textbf{F}_{2}(x, y) = 1 - 11x + 2x^{3} + 11y - 2y^{3}\\
  \end{aligned}
  \right.
\end{equation}$$

```python
Simulation2 = sdq.data.GenericSimulator()
Simulation2.set_initial_conditions_sampling_distribution("normal")
Simulation2.get_initial_conditions(
    loc=0, scale=1, size=[5000, 2], plot=True, n_bins=40
)  # i.e., normal(loc=0.0, scale=1.0, size=None)
Simulation2.create_time_vector(time_span=0.35, n_samples=100)
Simulation2.simulate_ODE("four_attractor_2d", plot_title="Four Attractor Simulation")
```
