```python
import scvelo as scv
import sklearn

adata = scv.datasets.pancreas()
adata = adata[:, adata.var["highly_variable_genes"] == "True"].copy()
adata.X = adata.X.toarray()

scaler = sklearn.preprocessing.StandardScaler()

X_scaled = scaler.fit_transform(adata.X)
adata.layers['X_scaled'] = X_scaled
adata
```

```python
model = sdq.scDiffEq(
    adata=adata,
    model="pancreas_demo",
    use_key="X_scaled",
    latent_dim=50,
    potential_type=None,
    time_cluster_key="clusters",
    t0_cluster="Ductal",
)
```

```python
model.fit(pretrain_epochs = 10, train_epochs = 10)
```
